# nn module(single neuron)

```python
#create model class
#5input + 1bias -> 1output
import torch
import torch.nn as nn

class Model(nn.Module):
  #inheriting the utilities from nn.module
  def __init__(self,num_features):
    super().__init__()
    self.lin = nn.Linear(num_features,1)#input features,output features
    self.sig = nn.Sigmoid()

  def forward(self,features):
    out = self.lin(features)
    out = self.sig(out)

    return out
```

```python
#create dataset
features = torch.rand(10,5)
#create model
model = Model(features.shape[1])

#call model for forward pass
model(features)#tp call forward pass it is a magic function
```

# nn.module using hidden layer

```python
class Model2(nn.Module):

  def __init__(self,num_features):
    super().__init__()
    self.linear = nn.Linear(num_features,3)
    self.rel = nn.ReLU()
    self.linear2 = nn.Linear(3,1)
    self.sig = nn.Sigmoid()

  def forward(self,features):

    out = self.linear(features)
    out = self.rel(out)
    out = self.linear2(out)
    out = self.sig(out)

    return out
```

```python
#create dataset
features = torch.rand(10,5)
#create model
model = Model2(features.shape[1])

#call model for forward pass
model(features)#tp call forward pass it is a magic function
```

# Sequential Containers

```python
class Model2(nn.Module):

  def __init__(self,num_features):
    super().__init__()
    self.network = nn.Sequential(
        nn.Linear(num_features,3),
        nn.ReLU(),
        nn.Linear(3,1),
        nn.Sigmoid()
    )

  def forward(self,features):

    out = self.network(features)

    return out
```

```python
#create dataset
features = torch.rand(10,5)
#create model
model = Model2(features.shape[1])

#call model for forward pass
model(features)#tp call forward pass it is a magic function
```
