```python
import time
size = 10000
matrix1 = torch.rand(size,size)
matrix2 = torch.rand(size,size)

start_time = time.time()
result_cpu=torch.matmul(matrix1,matrix2)
cpu_time=time.time()-start_time
print(f"Time on CPU: {cpu_time:.4f} seconds")

matrix1_gpu = matrix1.to('cuda')
matrix2_gpu = matrix2.to('cuda')

start_time = time.time()
result_gpu = torch.matmul(matrix1_gpu,matrix2_gpu)
torch.cuda.synchronize()
gpu_time = time.time() - start_time
print(f"time on gpu:{gpu_time:.4f} seconds")

#compare results
print(cpu_time/gpu_time)

```

```python
def binary_cross_entropy_loss(prediction,target):
  epislon = 1e-8
  prediction = torch.clamp(prediction,epislon,1-epislon)
  return -(target*torch.log(prediction)+(1-target)*torch.log(1-prediction))
```

```python
x = torch.tensor(6.7)
y = torch.tensor(0.0)
```

```python
w = torch.tensor(1.0,requires_grad = True)
b = torch.tensor(0.0,requires_grad = True)
```

```python
w
```

```python
b
```

```python
z = w*x + b
```

```python
z
```

```python
y_pred = torch.sigmoid(z)
```

```python
y_pred
```

```python
loss = binary_cross_entropy_loss(y_pred,y)
```

```python
loss
```

```python
loss.backward()
```

```python
print(w.grad)
```

```python
print(b.grad)
```
