```python
# Dataset defines how to access individual samples from data

# __len__() returns total number of samples
# __getitem__(idx) returns one sample (X[idx], y[idx])

# Dataset stores references to data (in RAM or file paths)
# it does NOT necessarily store full data in memory
```

```python
# DataLoader uses Dataset to load data in batches

# it provides:
# - batching (batch_size)
# - shuffling (shuffle=True)
# - iteration during training

# Dataset → single sample
# DataLoader → batch of samples
```

```python
from sklearn.datasets import make_classification
import torch
```

```python
X , y = make_classification(
    n_samples = 10,
    n_features = 2,
    n_informative = 2,
    n_redundant = 0,
    n_classes = 2,
    random_state = 42

)
```

```python
X
```

```python
y
```

```python
x=torch.tensor(X,dtype=torch.float32)
y=torch.tensor(y,dtype=torch.long)
```

```python
from torch.utils.data import Dataset , DataLoader
```

```python
class CustomDataset(Dataset):
  def __init__(self,features,labels):
    self.features = features
    self.labels = labels

  def __len__(self):
    return self.features.shape[0]

  def __getitem__(self,index):
    return self.features[index],self.labels[index]
```

```python
dataset = CustomDataset(X,y)
```

```python
len(dataset)
```

```python
dataset[2]
```

```python
dataloader = DataLoader(dataset,batch_size=2,shuffle=True)
```

```python
for batch_features,batch_labels in dataloader:
  print(batch_features)
  print(batch_labels)
  print('-----------------')
```
